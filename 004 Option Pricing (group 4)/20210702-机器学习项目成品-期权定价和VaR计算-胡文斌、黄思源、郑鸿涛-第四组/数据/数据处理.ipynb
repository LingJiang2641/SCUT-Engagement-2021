```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'

import numpy as np
import pandas as pd
```

```python
data1 = pd.read_csv('600276历史数据.csv')
data2 = pd.read_csv('600900历史数据.csv')
data3 = pd.read_csv('601318历史数据.csv')

data1.head()
data2.head()
data3.head()
```

```python
data11 = pd.merge(data1.iloc[:,0:2],data2.iloc[:,0:2],on="日期",how = "inner")
data11.head()


```

```python
data = pd.merge(data11.iloc[:,:],data3.iloc[:,0:2],on="日期",how = "inner")
data 
```

```python
dict1 = {"日期":"Date","收盘_x":"600276","收盘_y":"600900","收盘":"601318"}

data = data.rename(columns = dict1 )

data
```

```python
names = np.array(data.columns)[[1,2,3]]
# names = ["600276","600900","601318"]

for i in range(3):
    data[names[i]+"_Return"] = np.log(data[names[i]])-np.log(data[names[i]].shift(1))

data.head()
```

```python
data.to_csv('good_data.csv',index=False,header=True) 
```
